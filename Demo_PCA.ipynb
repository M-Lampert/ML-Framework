# Test of the PCA module

## 1. Import example dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

from framework.pca import PCA

%load_ext autoreload
%autoreload 2

iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
iris.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
iris.dropna(how="all", inplace=True)

plt.rcParams["figure.figsize"] = (12, 8)
plt.scatter(iris.sepal_length, iris.sepal_width)

X = iris.iloc[:, 0:4].values
```

## 2. Examine the data with the PCA module
### Eigenvalues and eigenvectors:

```python
eigen_values, eigen_vectors = PCA.get_eigendecomposition(X)
print(f"Eigenvalues:\n{eigen_values}")
print(f"Eigenvectors:\n{eigen_vectors}")
```

### Explained variance by the different features:

```python
# print(pca.get_variance_explained())
PCA.visualize_eigenvalues(X)
```

### Cumulative explained variance:
Shows that the first two features together explain 95.8% of the variance in the data.

```python
print(PCA.get_cumulative_variance_explained(X))
```

## 3. Project the data onto a lower-dimensional subspace
As the first two features together explain >95% of the variance in the data, we compute the PCA transformation
on the original data using features 0 and 1.

At first, we need to create a PCA object:

```python
pca = PCA.create_PCA(data=X, dimensions_or_variance=2)
pca.get_projection_matrix()
```

Then we use it for the transformation of the data;
either a whole batch of data like:

```python
X
```

```python
X_transformed = pca.transform(X)
X_transformed
```

Or for "single data":

```python
print(X[0])
pca.transform(X[0])
```

Alternatively you can directly choose the desired variance instead of the number of dimensions:

```python
pca2 = PCA.create_PCA(X, 0.95)
X_transformed = pca2.transform(X)
X_transformed
```
